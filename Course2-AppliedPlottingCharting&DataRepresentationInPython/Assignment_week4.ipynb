import pandas as pd
import matplotlib.pyplot as plt

#Load all data
data_18 = pd.read_excel('data_2018.xlsx')

#Filter by provinces (like a "states") on 2018 and 2017

acoruna_17 = data_17[data_17['CÓDIGO PROVINCIA'] == 15]

#Total of each category to plot a simple barchart
